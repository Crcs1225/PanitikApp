```python
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
df = pd.read_csv('Application Evaluation Software Engineering.csv')
```

```python
print(df.info())
```

```python
pattern = r'\bapp\b'
question_columns = [col for col in df.columns if re.search(pattern, col, flags=re.IGNORECASE)]
```

```python
print(question_columns)
```

```python
for i, column in enumerate(question_columns):
    plt.figure(figsize=(8, 6))

    # Plotting
    df[column].value_counts().plot(kind='bar', color='grey')

    # Adding Labels and Titles
    plt.title(f'Distribution of Responses for {column}')
    plt.xlabel('Response')
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility

    # Show Plot
    plt.tight_layout()  # Adjust layout to prevent clipping of labels
    plt.savefig(f'plot_{i + 1}.jpg')

    plt.show()
```
